```julia
using ODE
function f6(t, y)
    # Extract the components of the y vector
    (x, v) = y
    
    # Our system of differential equations
    x_prime = -x^2
    v_prime = x^2-v
    
    # Return the derivatives as a vector
    [x_prime; v_prime]
end;
```

```julia
# Initial condtions -- x(0) and x'(0)
const starty = [1.0; 0.0]

# Time vector going from 0 to 2*PI in 0.01 increments
time = 0:0.1:1;
```

```julia
t, y = ode45(f6, starty, time);
y1 = map(v -> v[1], y)
y2 = map(v -> v[2], y)
# print(y1)
```

```julia
using PyPlot
plot(t, y1, label="x(t)")
plot(t, y2, label="x'(t)")
legend();
```
